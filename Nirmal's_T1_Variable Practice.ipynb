```python
first_name = 'NIRMAL'
last_name = 'SHARMA'
country = 'INDIA'
city = 'INDORE'
age = 28
is_married = False
skills = ['Pyhton', 'DSA', 'AWS', 'Azure']
person_info = {
    'first_name' : 'NIRMAL',
    'last_name' : 'SHARMA',
    'country' : 'INDIA',
    'city' : 'INDORE',
    }

print ('First name:', first_name)
print ('First name length:' , len(first_name))
print ('Last name:', last_name)
print ('Last name length:', len(last_name))
print ('country:', country)
print ('City:', city)
print ('Age:', age)
print ('Married:', is_married)
print ('Skills:', skills)
print ('Person information:', person_info)
```
