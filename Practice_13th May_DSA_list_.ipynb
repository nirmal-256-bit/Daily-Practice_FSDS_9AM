# List Slicing 


```python
mylist = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight']
print(mylist)


print(mylist[0:3]) # Return all items from 0th to 3rd index location excluding the item

print(mylist[2:5]) #List all items from 2nd to 5th index location excluding the item a

print(mylist[:3])  #Return first three items
 
print(mylist[:2])  # Return first two items

print(mylist[-3:]) # Return last three items

print(mylist[-2:]) # Return last two items

print(mylist[-1])  # Return last item of the list

print(mylist[ : ]) # Return whole list
```

# all / any 

the all () method returns:

- True - if all elements in a list are true
- False - if any element in a list is false

The any() function returns True if any element in the list in True. if no, any() returns False. 



```python
L1 = [1,2,3,4,0]
```

```python
all(L1) # will return false as one value is false (value 0)
```

```python
any(L1) # will return true as we have items in the list with True value 
```

```python
L2 = [1,2,3,4,True,False]
```

```python
all(L2) # Returns false as one value is false 
```

```python
any(L2) #will Returns True as we have items in the list with True Value 
```

```python
L3 = [1,2,3,True]
```

```python
all(L3) # will return True as all items in the list are True
```
