# Number system 
 	    1. Binary Number system(bin)
		2. Octal Number System (oct)
		3. Decimal number system
        4. Hexadecimal(hex)

# 1. Binary number system (bin) 

only Value 1 & 2 

```python
# Decimal to Binary conversion 

print ('bin(55)=', bin(55)) #output : 0b110111


# Binary To decimal Converstion 

print ('int(0b110111)=',int(0b110111)) #output :55


#bool to binary Conversion 

print('bin(True)=',bin(True)) #output : 0b1
print('bin(False)=',bin(False)) #output : 0b0

# Complex to binary Conversion
print(bin(10+20j)) #output : TypeError--> 'complex' object cannot be interpreted as an integer
```

# 2. Octal Number System (oct) 

```python
#Decimal to Octal number System conversion 

print ('oct(45)=', oct(45)) #output:0o55

print ('int(0o55)=', int(0o55)) #output:45


#bool to Octal Number Conversion 

print('oct(True)=',oct(True)) #output : 0o1
print('oct(False)=',oct(False)) #output :0o0

```

# 3. Decimal number system (int)

Any number you type without a prefix is automatically treated as a decimal (base-10) value


```python
print('print(25)=', 25) #output : 25
```

# 4. Hexadecimal(hex)

```python
print('hex(95)=',hex(95)) #output : 0x5f

print ('int(0x5f)=',int(0x5f)) #output:95   

#bool to binary Conversion 

print('hex(True)=',hex(True)) #output : 0x1
print('hex(False)=',hex(False)) #output :0x0
```

# Bit wise operator
          1. Complement (~) 
		  2. Bitwise and (&)
		  3. Bitwise or (|)
		  4. xor (xclusive or) ^ 
		  5. left shift << 
          6. right shift >> 

# 1. Complement (~)

```python
print('~ 120 = ', ~120)  #output : -121
print('~(-120)= ', ~(-120)) #output : 119
```

#  2. Bitwise and (&)

Rule: output bit = 1 only if both input bits = 1

```python
print ('12 (0b1100) & 13 (0b1101) =', 12 & 13) #Output 12
```

# 3. Bitwise or (|)

```python
print(bin(25))
print(bin(65))


print ('25 | 65 = ', 25 | 65)

print(bin (89))

print('True | False=', True | False)
print('False | False=', False | False)
print('True | True =', True | True)
```

#  4. xor (xclusive or) ^



```python
print( bin(75))
print(bin(45))
 
Result = 75 ^ 45

print (' 75  ^ 45 =', Result)

print(bin(Result))
```

# 5. left shift ( << )
(zeros fill from right)


```python
n = 3 #0b11
result = n << 2
print('n<<2 =', result)

print(bin(result))
```

# 6. right shift (>>)
(bits fall off the right)

```python
n = 97 #0b1100001
result = n >> 2

print('n >> 2 =', result)

print(bin(result))
```
