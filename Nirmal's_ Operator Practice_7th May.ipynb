# Operator 


#there are 5 types of Operators in the Python 

1. Arithmetic [ +. -, /, //, %, %% , *, **, ]
2. Assignment [+=, -=, *= , /=, //=
3. Relational : Using this for Camparing [ >, <, <=, >=, ==, != ] 
4. Logical : [ and, or, not ] 
5. Unary : [- ]

# 1. Arithmetic Operators 

```python
x1, y1 = 10, 5

print('x1=', x1) # output x1= 10
print('y1=', y1) # output y1= 5

print( 'x1 + y1=', x1+y1 )
print( 'x1 - y1=', x1-y1 )
print( 'x1 * y1=', x1*y1 )
print( 'x1 / y1=', x1/y1 )
print( 'x1 % y1=', x1 % y1)
print( 'x1 ** y1=', x1**y1)
print( 'x1 ^ y1 =', x1 ^ y1)
```

# 3. Relational Operator

we are using this Operator for comparing 

 <, >, <=, >=, ==, != 

```python
print ('a =', a)
print ('b =', b)

print ('a > b =', a > b) #Out put: a > b = False
print ('a < b =', a < b) #Out put: a < b = True
print ('a <= b =', a <= b) #Out put: a <= b = True
print ('a >= b =', a >= b) #Out put: a >= b = False
print ('a != b =', a != b) #Out put: a != b = True
print ('a == b =', a == b) #Out put: a == b = False
```

# 'AND' 


```python
a = 5
b = 4

print('a =', a)
print('b =', b)
print('a<8 and b<5 =',a<8 and b<5)
print('a>8 and b>5 =',a>8 and b>5)
print('a<8 and b>5 =',a<8 and b>5)
print('a>8 and b<5 =',a>8 and b<5)
```

# 'OR'  

```python
print('a =', a)
print('b =', b)
print('a<8 or b<5 =',a<8 or b<5)
print('a>8 or b>5 =',a>8 or b>5)
print('a<8 or b>5 =',a<8 or b>5)
print('a>8 or b<5 =',a>8 or b<5)
```

# 'NOT'

```python
x = False 
print('x =',x)
print('not x=', not x)
```
